```python
import matplotlib.pyplot as plt
```

# Below code is for Matplotlib only and it helps to create inline plots

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace(-5,5,50)
y= x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Parabola')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented Method
fig=plt.figure() 
axes=fig.add_axes([0.1,0.1,0.6,0.8])
axes.plot(x,y)
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Parabola')
```

```python
#Why need to put this again
fig=plt.figure()
axes=fig.add_axes([0.1,0.2,0.8,1.0])
axes.plot(x,y)
axes.set_title('Big')
axes1= fig.add_axes([0.3,0.8,0.4,0.3])
axes1.plot(y,x)
axes1.set_title('Small')
```
