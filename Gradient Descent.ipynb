```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([3.7, 3.6, 2.8, 5.6, 3.3, 2.2, 3.3, 3.1, 3.2, 3.5,  5.2, 4.6,  5.8, 3])
y=np.array([5.7, 5.9, 6.7, 9.5, 5.4, 3.5, 6.2, 4.7, 6.1, 4.9, 10.7, 7.6, 11.8, 4.1])
```

```python
si=[0,0]
n=x.size
r=0.01
error=[]
```

```python
def costfunction(n,x,y,r,si):
    e=0
    for i in range(n):
        e+=(si[0]*x[i]+si[1]-y[i])**2
    e=e/(2*n)
    global error
    error.append(e)
    #print(e)
    pd_m=0
    pd_c=0
    for i in range(n):
        pd_m+=(si[0]*x[i]+si[1]-y[i])*x[i]
        pd_c+=(si[0]*x[i]+si[1]-y[i])
    pd_m=pd_m/n
    pd_c=pd_c/n
    si[0]=si[0]-(r*pd_m)
    si[1]=si[1]-(r*pd_c)
    #print(si[0])
    #print(si[1])
    return si
```

```python
for i in range(25000):
    global si
    si=costfunction(n,x,y,r,si)
```

```python
print('Slope of Best Fit Line is:')
print(si[0])
print('Intercept of Best Fit Line is:')
print(si[1])
print('Final Error is:')
print(error[24999])
```
